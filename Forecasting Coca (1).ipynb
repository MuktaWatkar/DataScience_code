```python
quarter=['Q1','Q2','Q3','Q4']
n=coca['Quarter'][0]
n
```

```python
coca['quarter']=0
for i in range(42):
    n=coca['Quarter'][i]
    coca['quarter'][i]=n[0:2]
    d=pd.DataFrame(pd.get_dummies(coca['quarter']))
    Coca=pd.concat((coca,d),axis=1)
```

```python
Coca
```

```python
t=np.arange(1,43)
Coca['t']=t
```

```python
t_sq=Coca['t']*Coca['t']
Coca['t_square']=t_sq
```

```python
Coca
```

```python
log=np.log(Coca['Sales'])
Coca['Log_sale']=log
```

```python
Coca
```

```python
plt.figure(figsize=(15,4))

sns.lineplot(x='quarter',y='Sales',data=Coca)
```

```python
Coca.shape
```

```python
Train=Coca.head(34)
Test=Coca.iloc[34:38,:]
pred=Coca.tail(4)
Cola=Coca.iloc[0:38,:]
```

```python
Train
```

```python
Test
```

```python
pred
```

```python
import statsmodels.formula.api as smf
LM=smf.ols("Sales~t",data=Train).fit()
predL=pd.Series(LM.predict(pd.DataFrame(Test['t'])))
rmse_LM=np.sqrt(np.mean((np.array(Test['Sales'])-np.array(predL))**2))
rmse_LM
            
                
```

```python
Exp = smf.ols('Log_sale~t',data=Train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
Quad = smf.ols('Sales~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
add_sea = smf.ols('Sales~Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
add_sea_Quad = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['Q1','Q2','Q3','Q4','t','t_square']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
Mul_sea = smf.ols('Log_sale~Q1+Q2+Q3+Q4',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(Test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
Mul_Add_sea = smf.ols('Log_sale~t+Q1+Q2+Q3+Q4',data = Train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(Test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea
```

```python
data = {"MODEL":pd.Series(["rmse_Linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_LM,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```
