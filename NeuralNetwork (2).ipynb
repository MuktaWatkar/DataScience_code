# Neural Network 
# Data=Forestfires & GasTurbines

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
forest=pd.read_csv("D:/ASSIGNMENTS/A16 Neural Network/forestfires.csv")
forest.head(2)
```

```python
forest.describe()
```

```python
forest.info()
```

```python
labelencoder=LabelEncoder()
forest['month']=labelencoder.fit_transform(forest['month'])
forest['day']=labelencoder.fit_transform(forest['day'])
forest['size_category']=labelencoder.fit_transform(forest['size_category'])
```

```python
forest
```

```python
forest.shape
```

```python
x=forest.iloc[:,0:11]
x
```

```python
y=forest.iloc[:,-1]
```

```python
y
```

```python
model=Sequential()
model.add(Dense(12,input_dim=11,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,validation_split=0.33, epochs=150,batch_size=10)
```

```python
scores=model.evaluate(x,y)
```

```python
print("%s : %2f%%" %(model.metrics_names[1],scores[1]*100))
```
